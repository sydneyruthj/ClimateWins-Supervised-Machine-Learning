#### Dataset is missing columns for ROMA_temp_min and ROMA_temp_max

```python
df_temp = df[columns]
df_temp.head()
```

```python
### 2.2 Check for missing values
df_temp.isnull().sum()
```

#### No nulls

### 2.3 Scaling

```python
#create scaler object using StandardScaler from sklearn.preprocessing
scaler = StandardScaler()
```

```python
#create new df with scaled data

df_temp_scaled = pd.DataFrame(scaler.fit_transform(df_temp), columns=columns)
df_temp_scaled.head()
```

## 3. Export Data
